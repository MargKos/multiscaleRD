```python
from __future__ import division
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import seaborn as sns
from ParametersB_SIR import *
import matplotlib.lines as mlines


SusJSDSimulations=np.load('./Solutions/JSDSusExplicitB.npy') 
RecJSDSimulations=np.load('./Solutions/JSDInfExplicitB.npy')
InfJSDSimulations=np.load('./Solutions/JSDRecExplicitB.npy')

#%%
#transfer to log() 
SusJSDSimulations=np.log(SusJSDSimulations)
RecJSDSimulations=np.log(RecJSDSimulations)
InfJSDSimulations=np.log(InfJSDSimulations)
#%%
Times=[4,9]
ts=0.1


sim_values = [1,10,25,50, 100, 150, 200, 250, 300, 350, 400, 450, 500]
```

```python

sim=500
# Plot for Susceptible
plt.figure(figsize=(9.3, 7.8))
plt.plot(sim_values,SusJSDSimulations[0:sim, 0], '--', color='blue', linewidth=4, label=f't = {np.round(ts*(Times[0]+1), decimals=2)}')
plt.plot(sim_values,SusJSDSimulations[0:sim, 1], ':', color='blue', linewidth=4, label=f't = {np.round(ts*(Times[1]+1), decimals=2)}')

plt.plot(sim_values,SusJSDSimulations[0:sim, 0], 'o', color='blue', markersize=10)
plt.plot(sim_values,SusJSDSimulations[0:sim, 1], 'o', color='blue', markersize=10)

plt.xlabel('Number of Simulations', fontsize=40)
plt.ylabel('log(JS-divergence)', fontsize=40)

# Set x-axis ticks to be integers only
xticks = range(0,500, 100)  # Adjust according to your data limits
plt.xticks(xticks, fontsize=30)

# Set y-axis ticks to be integers only
yticks = range(-7,2, 1)  # Adjust based on your y-axis limits
plt.yticks(yticks, fontsize=30)

plt.ylim(-7, 2)
plt.xlim(0, 500)

# Custom legend
blue_dashed = mlines.Line2D([], [], color='black', linestyle='--', linewidth=5, label=f't = {np.round(ts*(Times[0]+1), decimals=2)}')
blue_dotted = mlines.Line2D([], [], color='black', linestyle=':', linewidth=5, label=f't = {np.round(ts*(Times[1]+1), decimals=2)}')
blue_line = mlines.Line2D([], [], color='blue', marker='o', linestyle='None', markersize=20, label='S')
plt.legend(handles=[blue_dashed, blue_dotted, blue_line], fontsize=40,  loc='upper left', bbox_to_anchor=(0.43, 1.03))
plt.grid(which='both', linestyle='--', linewidth=0.7, color='gray')  # Customize as needed

plt.tight_layout()
plt.savefig('./Plots/Explicit/JSD_Susceptible_ExplicitB.png')


# Plot for Infected
plt.figure(figsize=(9.3, 7.8))
plt.plot(sim_values,InfJSDSimulations[0:sim, 0], '--', color='red', linewidth=4, label=f't = {np.round(ts*(Times[0]+1), decimals=2)}')
plt.plot(sim_values,InfJSDSimulations[0:sim, 1], ':', color='red', linewidth=4, label=f't = {np.round(ts*(Times[1]+1), decimals=2)}')

plt.plot(sim_values,InfJSDSimulations[0:sim, 0], 'o', color='red', markersize=10)
plt.plot(sim_values,InfJSDSimulations[0:sim, 1], 'o', color='red', markersize=10)

plt.xlabel('Number of Simulations', fontsize=40)
plt.ylabel('log(JS-divergence)', fontsize=40)


# Set x-axis ticks to be integers only
xticks = range(0,500, 100)  # Adjust according to your data limits
plt.xticks(xticks, fontsize=30)

# Set y-axis ticks to be integers only
yticks = range(-7,2, 1)  # Adjust based on your y-axis limits
plt.yticks(yticks, fontsize=30)

plt.ylim(-7, 2)
plt.xlim(0, 500)
# Custom legend
red_line = mlines.Line2D([], [], color='red', marker='o', linestyle='None', markersize=20, label='I')
plt.legend(handles=[blue_dashed, blue_dotted, red_line], fontsize=40,  loc='upper left', bbox_to_anchor=(0.43, 1.03))
plt.grid(which='both', linestyle='--', linewidth=0.7, color='gray')  # Customize as needed

plt.tight_layout()
plt.savefig('./Plots/Explicit/JSD_Infected_ExplicitB.png')


# Plot for Recovered
plt.figure(figsize=(9.3, 7.8))
plt.plot(sim_values,RecJSDSimulations[0:sim, 0], '--', color='green', linewidth=4, label=f't = {np.round(ts*(Times[0]+1), decimals=2)}')
plt.plot(sim_values,RecJSDSimulations[0:sim, 1], ':', color='green', linewidth=4, label=f't = {np.round(ts*(Times[1]+1), decimals=2)}')

plt.plot(sim_values,RecJSDSimulations[0:sim, 0], 'o', color='green', markersize=10)
plt.plot(sim_values,RecJSDSimulations[0:sim, 1], 'o', color='green', markersize=10)

plt.xlabel('Number of Simulations', fontsize=40)
plt.ylabel('log(JS-divergence)', fontsize=40)
# Set x-axis ticks to be integers only
xticks = range(0,500, 100)  # Adjust according to your data limits
plt.xticks(xticks, fontsize=30)

# Set y-axis ticks to be integers only
yticks = range(-7,2, 1)  # Adjust based on your y-axis limits
plt.yticks(yticks, fontsize=30)

plt.ylim(-7, 2)
plt.xlim(0, 500)

# Custom legend
green_line = mlines.Line2D([], [], color='green', marker='o', linestyle='None', markersize=20, label='R')
plt.legend(handles=[blue_dashed, blue_dotted, green_line], fontsize=40, loc='upper left', bbox_to_anchor=(0.43, 1.03))

plt.grid(which='both', linestyle='--', linewidth=0.7, color='gray')  # Customize as needed

plt.tight_layout()
plt.savefig('./Plots/Explicit/JSD_Recovered_ExplicitB.png')


```
