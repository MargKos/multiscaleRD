```python
from __future__ import division
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import seaborn as sns
from Parameters_LV import *
import matplotlib.lines as mlines

# JSD for LV from Paper (https://arxiv.org/abs/2006.00003)

PreyJSDSimulations=np.load('./Solutions/LVPreyJSDPaper.npy') 
PredJSDSimulations=np.load('./Solutions/LVPredJSDPaper.npy')

#%%
#transfer to log() if necessary
PreyJSDSimulations=np.log(PreyJSDSimulations)
PredJSDSimulations=np.log(PredJSDSimulations)

```

```python
# Define times and time step
Times=[39,69,89]
ts = 0.1

# Plot JSD Preys

plt.figure(figsize=(9.3, 7.8))
plt.plot(PreyJSDSimulations[0:500, 1], '--', color='blue', linewidth=4, label=f't = {ts*(Times[1]+1)}')
plt.plot(PreyJSDSimulations[0:500, 2], ':', color='blue', linewidth=4, label=f't = {ts*(Times[2]+1)}')
plt.xlabel('Number of Simulations', fontsize=40)
plt.ylabel('log(JS-divergence)', fontsize=40)

# Set x-axis ticks to be integers only
xticks = range(0,500, 100)  # Adjust according to your data limits
plt.xticks(xticks, fontsize=30)

# Set y-axis ticks to be integers only
yticks = range(-8,-1, 1)  # Adjust based on your y-axis limits
plt.yticks( fontsize=30)

plt.ylim(-8, -1)
plt.xlim(0, 500)

# Custom legend
blue_dashed = mlines.Line2D([], [], color='black', linestyle='--', linewidth=5, label=f't = {ts*(Times[1]+1)}')
blue_dotted = mlines.Line2D([], [], color='black', linestyle=':', linewidth=5, label=f't = {ts*(Times[2]+1)}')
blue_line = mlines.Line2D([], [], color='blue', marker='o', linestyle='None', markersize=20, label='Prey')
plt.legend(handles=[blue_dashed, blue_dotted, blue_line], fontsize=40,  loc='upper left', bbox_to_anchor=(0.43, 1.03))
plt.grid(which='both', linestyle='--', linewidth=0.7, color='gray')  # Customize as needed

plt.tight_layout()
plt.savefig('./Plots/JSD_PreyPaper.png')


# Plot JSD Predators
plt.figure(figsize=(9.3, 7.8))
plt.plot(PredJSDSimulations[0:500, 1], '--', color='red', linewidth=4, label=f't = {ts*(Times[1]+1)}')
plt.plot(PredJSDSimulations[0:500, 2], ':', color='red', linewidth=4, label=f't = {ts*(Times[2]+1)}')
plt.xlabel('Number of simulations', fontsize=40)
plt.ylabel('log(JS-divergence)', fontsize=40)


# Set x-axis ticks to be integers only
xticks = range(0,500, 100)  # Adjust according to your data limits
plt.xticks(xticks, fontsize=30)

# Set y-axis ticks to be integers only
yticks = range(-8,-1, 1)  # Adjust based on your y-axis limits
plt.yticks(fontsize=30)

plt.ylim(-8, -1)
plt.xlim(0, 500)
# Custom legend
red_line = mlines.Line2D([], [], color='red', marker='o', linestyle='None', markersize=20, label='Pred.')
plt.legend(handles=[blue_dashed, blue_dotted, red_line], fontsize=40,  loc='upper left', bbox_to_anchor=(0.43, 1.03))
plt.grid(which='both', linestyle='--', linewidth=0.7, color='gray')  # Customize as needed

plt.tight_layout()
plt.savefig('./Plots/JSD_PredPaper.png')


```
