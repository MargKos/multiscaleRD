```python
from __future__ import division
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import seaborn as sns
from ParametersB_SIR import *
import matplotlib.lines as mlines


SusJSDSimulationsExplicit=np.load('./Solutions/JSDSusExplicitB.npy') 
RecJSDSimulationsExplicit=np.load('./Solutions/JSDInfExplicitB.npy')
InfJSDSimulationsExplicit=np.load('./Solutions/JSDRecExplicitB.npy')

SusJSDSimulationsTau=np.load('./Solutions/JSDSusTauB.npy') 
RecJSDSimulationsTau=np.load('./Solutions/JSDInfTauB.npy')
InfJSDSimulationsTau=np.load('./Solutions/JSDRecTauB.npy')


#%%

Times=[4,9]
ts=0.1


sim_values = [1,10,25,50, 100, 150, 200, 250, 300, 350, 400, 450, 500]
```

```python

SusJSDSimulationsTau = np.log(SusJSDSimulationsTau)
RecJSDSimulationsTau = np.log(RecJSDSimulationsTau)
InfJSDSimulationsTau = np.log(InfJSDSimulationsTau)



SusJSDSimulationsExplicit = np.log(SusJSDSimulationsExplicit)
RecJSDSimulationsExplicit = np.log(RecJSDSimulationsExplicit)
InfJSDSimulationsExplicit  = np.log(InfJSDSimulationsExplicit)
```

```python
sim=500
time=0
# Plot for Susceptible
plt.figure(figsize=(9.3, 7.8))
plt.plot(sim_values,SusJSDSimulationsTau[0:sim, time], '--', color='blue', linewidth=4, label='Tau', decimals=2)}')
plt.plot(sim_values,SusJSDSimulationsExplicit[0:sim, time], ':', color='blue', linewidth=4, label='Explicit', decimals=2)}')

plt.plot(sim_values,SusJSDSimulationsTau[0:sim, time], 'o', color='blue', markersize=10)
plt.plot(sim_values,SusJSDSimulationsExplicit[0:sim, time], 'o', color='blue', markersize=10)

plt.xlabel('Number of Simulations', fontsize=40)
plt.ylabel('log(JS-divergence)', fontsize=40)

# Set x-axis ticks to be integers only
xticks = range(0,500, 100)  # Adjust according to your data limits
plt.xticks(xticks, fontsize=30)

# Set y-axis ticks to be integers only
yticks = range(-7,2, 1)  # Adjust based on your y-axis limits
plt.yticks(yticks, fontsize=30)

plt.ylim(-7, 2)
plt.xlim(0, 500)

# Custom legend
blue_dashed = mlines.Line2D([], [], color='black', linestyle='--', linewidth=5, label=f't = {np.round(ts*(Times[0]+1), decimals=2)}')
blue_dotted = mlines.Line2D([], [], color='black', linestyle=':', linewidth=5, label=f't = {np.round(ts*(Times[1]+1), decimals=2)}')
blue_line = mlines.Line2D([], [], color='blue', marker='o', linestyle='None', markersize=20, label='S')
plt.legend(handles=[blue_dashed, blue_dotted, blue_line], fontsize=40,  loc='upper left', bbox_to_anchor=(0.43, 1.03))
plt.grid(which='both', linestyle='--', linewidth=0.7, color='gray')  # Customize as needed

plt.tight_layout()
plt.savefig('./Plots/JSD_Susceptible_Comparison.png')

```

```python

# Differenz berechnen (Explicit - Tau)
SusJSD_Diff = SusJSDSimulationsExplicit - SusJSDSimulationsTau
RecJSD_Diff = RecJSDSimulationsExplicit - RecJSDSimulationsTau
InfJSD_Diff = InfJSDSimulationsExplicit - InfJSDSimulationsTau

# Iterationen als x-Achse
iterations = np.arange(1, len(SusJSD_Diff) + 1)



# Plot erstellen
plt.figure(figsize=(10, 8))
plt.plot(sim_values , SusJSD_Diff[:, time], label="Sus. JSD Diff (Explicit - Tau)", linestyle="-", color='blue', marker="o", markersize=10)
plt.plot(sim_values , RecJSD_Diff[:, time], label="Rec. JSD Diff (Explicit - Tau)", linestyle="-", color='green', marker="s", markersize=10)
plt.plot(sim_values , InfJSD_Diff[:, time], label="Inf. JSD Diff (Explicit - Tau)", linestyle="-", color='red', marker="^", markersize=10)

plt.text(sim_values [-5], 0.01, "Tau is better", color="black", fontsize=20, verticalalignment="bottom")
plt.text(sim_values [-5], -0.1, "Explicit is better", color="black", fontsize=20, verticalalignment="top")

# Null-Linie zur Referenz
plt.axhline(y=0, color="black", linestyle="--", linewidth=1)
plt.yticks([-0.1, -0.05, 0, 0.05, 0.1], fontsize=30)
# Labels und Titel
plt.xlabel("Iteration",fontsize=30)
plt.ylabel("Difference of JS-Divergence",fontsize=30 )
plt.xticks(fontsize=30)
plt.legend(fontsize=20)
plt.grid(True, linestyle="--", alpha=0.6)
#plt.savefig('')
```
