```python
from __future__ import division
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import seaborn as sns
from Parameters_Proliferation import *
import matplotlib.lines as mlines
#import animation

PreyJSDSimulations=np.load('./Solutions/JSD_Proliferation.npy') 

#%%
#transfer to log() 
PreyJSDSimulations_log=np.log(PreyJSDSimulations)

#%%
Times=[4,9,14]
ts=0.1

```

```python
# Create a new figure with 1 subplot (since you're plotting on a single axis)
fig, ax = plt.subplots(1, 1, figsize=(15, 15))

# Plot for Susceptible on the single axis
ax.plot(PreyJSDSimulations[0:500, 0], '-', color='blue', linewidth=3, label=f't = {ts*(Times[0]+1)}')
ax.plot(PreyJSDSimulations[0:500, 1], '--', color='blue', linewidth=3, label=f't = {ts*(Times[1]+1)}')
ax.plot(PreyJSDSimulations[0:500, 2], ':', color='blue', linewidth=3, label=f't = {ts*(Times[2]+1)}')

# Set labels and tick parameters for the single axis
ax.set_xlabel('Number of simulations', fontsize=25)
ax.set_ylabel('log(JS-divergence)', fontsize=25)
ax.tick_params(axis='both', which='major', labelsize=25)

# Combine all custom legend handles and add them to the single axis
blue_solid = mlines.Line2D([], [], color='black', linestyle='-', linewidth=3, label=f't = {ts*(Times[0]+1)}')
blue_dashed = mlines.Line2D([], [], color='black', linestyle='--', linewidth=3, label=f't = {ts*(Times[1]+1)}')
blue_dotted = mlines.Line2D([], [], color='black', linestyle=':', linewidth=3, label=f't = {ts*(Times[2]+1)}')

# Add the combined custom legend to the axis
ax.legend(handles=[blue_solid, blue_dashed, blue_dotted], fontsize=25, loc='upper right')

# Save the figure
plt.tight_layout()
plt.savefig('./Plots/JSD_Proliferation.png')
plt.show()
```
