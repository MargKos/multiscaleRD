```python
from __future__ import division
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import seaborn as sns
from Parameters_Proliferation import *
import matplotlib.lines as mlines
#import animation

PreyJSDSimulations=np.load('./Solutions/JSD_Proliferation.npy') 

#%%
#transfer to log() 
PreyJSDSimulations_log=np.log(PreyJSDSimulations)

#%%
Times=[39,69,89]
sim_values = [50,100,200,300,400, 500, 600,700,800,900,1000, 1100,1200]
ts=0.1

sim=1200
```

```python

# Plot setup
fig, ax = plt.subplots(1, 1, figsize=(15, 15))

# Plot for the first and third time frames (standard colors)
ax.plot(sim_values[0:sim], PreyJSDSimulations[0:sim, 0], 'o-', color='blue', markersize=15, linewidth=4, label=f't = {ts * (Times[0] + 1)}')
ax.plot(sim_values[0:sim], PreyJSDSimulations[0:sim, 2], 'o-', color='green', markersize=15, linewidth=4, label=f't = {ts * (Times[2] + 1)}')

# Plot for the second time frame (red dots, orange line)
ax.plot(sim_values[0:sim], PreyJSDSimulations[0:sim, 1], 'o-', color='orange', markerfacecolor='red', markersize=15, linewidth=4, label=f't = {ts * (Times[1] + 1)}')

# Set labels and ticks
ax.set_xlabel('Number of simulations', fontsize=45)
ax.set_ylabel('log(JS-divergence)', fontsize=45)
ax.tick_params(axis='both', which='major', labelsize=45)

# Add legend with custom handles
blue_line = mlines.Line2D([], [], color='blue', linestyle='-', marker='o', markersize=10, linewidth=4, label=f't = {ts * (Times[0] + 1)}')
orange_line_red_dots = mlines.Line2D([], [], color='orange', linestyle='-', marker='o', markerfacecolor='red', markersize=10, linewidth=4, label=f't = {ts * (Times[1] + 1)}')
green_line = mlines.Line2D([], [], color='green', linestyle='-', marker='o', markersize=10, linewidth=4, label=f't = {ts * (Times[2] + 1)}')

# Add the legend
ax.legend(handles=[blue_line, orange_line_red_dots, green_line], fontsize=45, loc='upper right')

# Save the figure
plt.tight_layout()
plt.savefig('./Plots/Proliferation/JSD_Proliferation.png')
plt.show()

```
